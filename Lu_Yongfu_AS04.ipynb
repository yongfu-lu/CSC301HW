```python
import numpy as np
import math 
```

```python
def f(x):
    return np.sin(x)
```

```python
# using np.cos(x)
x = 1
exact_der =np.cos(x)
print(exact_der)
```

```python
# vary of h
h_values = [0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.0000001, 0.000000001, 0.0000000001]
```

```python
#using three-point endpoint formula
deris = []
abs_errors =[]
re_errors = []
for h in h_values:
    deri = (-3*f(x) + 4*f(x+h) - f(x+2*h)) / (2*h)
    abs_error = abs(exact_der - deri)
    re_error = abs_error / exact_der
    
    deris.append(deri)
    abs_errors.append(abs_error)
    re_errors.append(re_error)
```

## Result of using three-point endpoint formula

```python
print("Value of h  |       derivative      |      absolute error      |     relative error")
for i in range(len(h_values)):
    print(h_values[i], end="           ")
    print(deris[i], end="       ")
    print(abs_errors[i], end = "       ")
    print(re_errors[i])
```

```python
# using five-point-midpoint formula
deris = []
abs_errors =[]
re_errors = []
for h in h_values:
    deri = (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x+2*h)) / (12*h)
    abs_error = abs(exact_der - deri)
    re_error = abs_error / exact_der
    
    deris.append(deri)
    abs_errors.append(abs_error)
    re_errors.append(re_error)
```

## Result of using five-point midpoint formula

```python
print("Value of h  |       derivative      |      absolute error      |     relative error")
for i in range(len(h_values)):
    print(h_values[i], end="           ")
    print(deris[i], end="       ")
    print(abs_errors[i], end = "       ")
    print(re_errors[i])
```

In this program, we assume that the result of np.cos(x) is the exact derivative value of sin(x). Also we use three-point endpoint method and five-point midpoint method to estimate the derivative of sin(x). We use different value of h to find the best result..

By comparing the result. We can conclude that when we use three-point endpoint method,  h = 1e-05 give the best resylt.

When we use five-point midpoint method, h = 0.0001 gives the best result.

Five-point midpoint has the better accuracy than three-point endpoint in general.
